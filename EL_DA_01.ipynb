# **Netflix Movies and TV Shows Data Set**

### **Import Required Libraries**

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### **Load(import) Data Using Pandas**

```python
data=pd.read_csv('netflix_titles.csv')
df=pd.DataFrame(data)
df
```

## **Data Cleaning**

**Step-1**:Check is there any null or duplicated values is there or not

```python
duplicate=df.duplicated().sum()
null=df.isnull().sum()
print("Duplicates:",duplicate)
print("Null Values:",null)
```

***So there is no duplicate values**                                
***And there are Null values in director column and cast column , country , date_added column and rating , duration column**

#### we don't know cast and director and country details so i will fill with unknown

```python
df['director'].fillna('unknown')
df['cast'].fillna('unknown')
df['country'].fillna('unknown')
null=df.isnull().sum() #to check the null values replaced or not
print(null)
```

### Know Fill Remaining all with mode

```python
df.fillna(df['date_added'].mode()[0],inplace=True) #it means it will fill with first mode value
df.fillna(df['rating'].mode()[0],inplace=True)
df.fillna(df['duration'].mode()[0],inplace=True)
null=df.isnull().sum()
print(null)
```

### **Cleaned the null and duplicates data**

### Know  Standardize text values like country names

```python
df['title']=df['title'].str.strip().str.title()
df['country'] = df['country'].str.strip().str.title()
df.head(5)
```

### Check the Data Types

```python
df.info()
```

### Convert date formats to a consistent type (e.g., dd-mm-yyyy)

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.info()
```

```python
df.head(3)
```

**Date Column Changed Successfully Changed**

###  Renaming column headers to be clean and uniform (e.g., lowercase, no spaces).

```python
df.columns = df.columns.str.strip().str.lower().str.replace('_','')
df.head(2)
```
